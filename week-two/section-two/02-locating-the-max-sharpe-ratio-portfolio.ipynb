#### We want to get the portfolio that gives us the Maximum Sharpe Ratio (MSR)

```python
from scipy.optimize import minimize

def msr(riskfree_rate, expected_return, cov):
    """
    Riskfree rate + Expected Return + COV matrix -> w (Weight vector)
    """
    n = expected_return.shape[0]
    init_guess = np.repeat(1/n, n)
    bounds = ((0.0, 1.0),) * n

    # constraints #1
    weights_sum_to_1 = {
        'type': 'eq',
        'fun': lambda weights: np.sum(weights) - 1
    }

    def negative_sharpe_ratio(weights, riskfree_rate, expected_return, cov):
        """
        Returns the negative of the sharep ratio, given weights
        """
        portfolio_return = erk.portfolio_return(weights, expected_return)
        portfolio_volatility = erk.portfolio_volatility(weights, cov)

        return -(portfolio_return - riskfree_rate) / portfolio_volatility

    results = minimize(
        negative_sharpe_ratio, init_guess,
        args=(riskfree_rate, expected_return, cov,), method="SLSQP", options={"disp": False},
        constraints=(weights_sum_to_1),
        bounds=bounds
    )

    return results.x
```

```python
ax = erk.plot_efficient_frontier(20, expected_return=expected_return, covariance_matrix=cov)

ax.set_xlim(left=0)
```

```python
ax = erk.plot_efficient_frontier(20, expected_return=expected_return, covariance_matrix=cov)

ax.set_xlim(left=0)

# 10% risk free rate
riskfree_rate = 0.1

weights_msr = msr(riskfree_rate, expected_return, cov)
return_msr = erk.portfolio_return(weights=weights_msr, returns=expected_return)
volatility_msr = erk.portfolio_volatility(weights=weights_msr, covariance_matrix=cov)

# add CML
cml_x = [0, volatility_msr]
cml_y = [riskfree_rate, return_msr]

ax.plot(cml_x, cml_y, color="green", marker="*", linestyle="dashed")
```
